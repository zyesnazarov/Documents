```python
import pandas as pd
!pip install datasets
from datasets import load_dataset
import matplotlib.pyplot as plt

#loading data

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data cleanup

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
job_count = df.job_title_short.value_counts().sort_values(ascending = False)
```

```python
plt.bar(job_count.index, job_count)
plt.title('Posting by job title')
plt.ylabel('Count by job posting')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

```python
job_count.plot(kind = 'bar')
plt.title('Posting by job title')
plt.ylabel('Count by job posting')
plt.xlabel('')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

```python
df.groupby(['job_title_short'])['salary_year_avg'].mean().sort_values()
```

```python
job_salary = df.groupby(['job_title_short'])['salary_year_avg'].median().sort_values()
```

```python
job_salary.plot(kind = 'barh')
plt.xlabel('Salary in USD')
plt.ylabel('Job title description')
plt.title('Median salary by job title')
```
